# PRE-PROCESSING

```python
#Eliminamos el customer id y churn (ya que esta columna es el valor que predeciremos)
del(X_pred['Customer_ID'])
del(X_pred['churn'])
```

```python
#Comprobamos que no hay duplicados
X_pred.duplicated().sum()
```

Importamos los diccionarios, las funciones utilizadas en el preprocessing y las lista de columnas eliminadas por baja varianza y alta correlación:

```python
dict_input_media= pickle.load(open(PATH +'dict_input_media.pkl','rb'))
dict_input_moda= pickle.load(open(PATH +'dict_input_moda.pkl','rb'))
cols_to_remove= pickle.load(open(PATH +'cols_to_remove.pkl','rb'))
```

```python
imputacion_nulos= pickle.load(open(PATH +'imputacion_nulos.pkl','rb'))
del_valores_negativos= pickle.load(open(PATH +'del_valores_negativos.pkl','rb'))
new_column_kids= pickle.load(open(PATH +'new_column_kids.pkl','rb'))
cat_columns_to_num= pickle.load(open(PATH +'cat_columns_to_num.pkl','rb'))
num_to_bool= pickle.load(open(PATH +'num_to_bool.pkl','rb'))
```

```python
#Aplicamos el mismo preprocesamiento hecho al dataset train-test
def prepro(df,dict_input_media,dict_input_moda):
    imputacion_nulos(df, dict_input_media, dict_input_moda)
    del_valores_negativos(df)
    new_column_kids(df)
    cat_columns_to_num(df)
    num_to_bool(df)
```

```python
prepro(X_pred,dict_input_media, dict_input_moda)
```

```python
#Comprobamos que no hay nulos en el dataset para la predicción.
X_pred.isnull().sum().sum()
```

```python
X_pred.info()
```

Tenemos un dataframe para predecir sin nulos, aplicamos OneHotEncoding

```python
X_pred_enc = pd.get_dummies(X_pred, columns=['new_cell','crclscod','asl_flag','prizm_social_one','area','dualband','refurb_new','hnd_webcap','ownrent','dwlltype','marital','infobase','ethnic',
 'creditcd'])
```

A este dataframe le tenemos que eliminar las col. correlacionadas y de baja varianza

```python
for i in X_pred_enc.columns:
    if i in cols_to_remove:
        del(X_pred_enc[i])
```

Tenemos que asegurarnos que las columnas con las que se ha entrenado el modelo son las mismas que tenemos en el dataset para la predicción:

```python
features_model =  list(best_model.feature_names_in_)
print('Estas son las variables que tenemos que conseguir en el preprocessing para que el modelo funcione:\n', features_model)
```

```python
features_test = list(X_pred_enc.columns)

print('Columnas en dataset:',len(features_test))
print('Variables en modelos:',len(features_model))
print('¿Match?:', features_model == features_test)

```

Comprobamos que columnas tenemos en el dataset pero que no son necesarias para la predicción.

```python
no_necesitamos=[]

for i in features_test:
    if i in features_model:
        continue
    else:
        no_necesitamos.append(i)
no_necesitamos
```

Tenemos dos columnas "nuevas", dos columnas que surgen de OneHotEncoding porque tenemos categorías nuevas que no habiamos visto, por lo tanto las borraremos.

```python
del(X_pred_enc['infobase_N'])
del(X_pred_enc['ethnic_C'])
```

```python
features_test = list(X_pred_enc.columns)

print('Columnas en dataset:',len(features_test))
print('Variables en modelos:',len(features_model))
print('¿Match?:', features_model == features_test)
```

Como el modelo utilizado es un modelo de reglas no es necesario escalar los datos.

## Predicción:

```python
predictions = best_model.predict(X_pred_enc)
predictions
```

```python
#Guardamos nuestras predicción en un dataframe.
predictions = pd.DataFrame(predictions, columns=['churn'], index = X_pred_enc.index)
```

```python
df_final_prediction= pd.merge(
    left= Customer_ID_index,
    right= predictions,
    left_index=True,
    right_index=True
)
```
